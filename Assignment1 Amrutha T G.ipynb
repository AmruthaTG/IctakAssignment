##### Pairplot between SL SW PL PW

```python
data = df[['SL','SW','PL','PW']]
sns.pairplot(data)
plt.show()
```

```python
sns.countplot(x="Classification", data=df)
plt.title("Classification Distribution")
plt.xlabel("Classification")
plt.xticks(rotation= 60)
plt.show()
```

```python
sns.boxplot(y =df['SL'], x=df['Classification'])
plt.title('Boxplot SL vs Classification')
plt.show()
```

```python
sns.boxplot(y =df['SW'], x=df['Classification'])
plt.title('Boxplot SW vs Classification')
plt.show()
```

```python
sns.boxplot(y =df['PL'], x=df['Classification'])
plt.title('Boxplot PL vs Classification')
plt.show()
```

```python
sns.boxplot(y =df['PW'], x=df['Classification'])
plt.title('Boxplot PW vs Classification')
plt.show()
```
